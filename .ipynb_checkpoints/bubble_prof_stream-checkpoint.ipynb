```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv("streamplot_data.csv")

# Sort values to ensure consistent axis ordering
df_sorted = df.sort_values(by=["v", "xi"])  # since xi is outer loop, v is inner

# Extract unique values for meshgrid axes
# xi_vals = np.sort(df_sorted["xi"].unique())
# v_vals = np.sort(df_sorted["v"].unique())
xi_vals = np.linspace(0.01, 0.99, 30)
v_vals = np.linspace(0.01, 0.99, 30)

# Create pivot tables to reshape the data into 2D grids
DXI = df_sorted.pivot(index="v", columns="xi", values="dxidtau").values
DV = df_sorted.pivot(index="v", columns="xi", values="dvdtau").values

# Meshgrid for plotting (ensure alignment with pivot table)
XI, V = np.meshgrid(xi_vals, v_vals)

# Normalize vectors for plotting
mag = np.sqrt(DV**2 + DXI**2)
mag[mag == 0] = 1  # Avoid division by zero
DV /= mag
DXI /= mag

# Phase space plot with trajectories
# NOTE: Integration direction here is inwards towards center of bubble, so
plt.figure(figsize=(8, 6))
plt.streamplot(XI, V, DXI, DV, density=0.3, color='k', linewidth=1, broken_streamlines=False)
# Labels and formatting
plt.xlabel(r"$\xi$")
plt.ylabel(r"$v(\xi)$")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid()

plt.show()
```
